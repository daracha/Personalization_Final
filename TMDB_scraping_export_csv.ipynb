```python
def movie_data(url,movieId):
    import requests
    from bs4 import BeautifulSoup
    response = requests.get(url)
    results_page = BeautifulSoup(response.content,'lxml')

    title=results_page.find('meta',{'property':'og:title'}).get('content')

    user_score=results_page.find('div',{'class':'user_score_chart'}).get('data-percent')

    featured_crew=[]
    for i in results_page.find_all('li',{'class':'profile'}):
        try:
            featured_crew.append((i.find('a').get_text(),(i.find('a').get('href')).split('/')[2].split('-')[0],i.find('p',{'class':'character'}).get_text()))
        except:
            pass

    top_billed_cast=[]
    for i in results_page.find_all('li',{'class':'card'}):
        try:
            top_billed_cast.append((i.find('p').get_text(),i.find('a').get('href').split('/')[2].split('-')[0],i.find('p',{'class':'character'}).get_text()))
        except:
            pass

    for i in results_page.find_all('strong'):
        if i.get_text()=='Original Language':
            original_language=i.parent.get_text().split()[2]

    for i in results_page.find_all('strong'):
        if i.get_text()=='Runtime':
            if (i.parent.get_text().split()[1]) != "-":
                runtime=(i.parent.get_text().split()[1])+(i.parent.get_text().split()[2])
            else:
                runtime="-"

    for i in results_page.find_all('strong'):
        if i.get_text()=='Budget':
            budget=i.parent.get_text().split()[1]

    for i in results_page.find_all('strong'):
        if i.get_text()=='Revenue':
            revenue=i.parent.get_text().split()[1]

    genres=[]
    for i in results_page.find('section',{'class':'genres right_column'}).find_all('a'):
        genres.append(i.get_text())

    all_features=[movieId,title,user_score,featured_crew,top_billed_cast,original_language,runtime,budget,revenue,genres]

    import pandas as pd
    pd.set_option('display.max_colwidth', -1)
    df=pd.DataFrame(all_features).transpose()
    df.columns=['movieId','TITLE','USER SCORE','FEATURED CREW','TOP BILLED CAST','ORIGINAL LANGUAGE','RUN TIME','BUDGET','REVENUE','GENRE']
    return df  
```
