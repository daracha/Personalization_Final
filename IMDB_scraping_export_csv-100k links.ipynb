```python
import pandas as pd
df=pd.read_csv('links_100k.csv')
df['imdbId'] = df['imdbId'].apply(lambda x: '{0:0>7}'.format(x))
df['links']=("http://www.imdb.com/title/tt"+df['imdbId'].astype(str))
df
```

```python
def movie_data(url,movieId,imdbId):
    import requests
    from bs4 import BeautifulSoup
    response = requests.get(url)
    results_page = BeautifulSoup(response.content,'lxml')

    title=results_page.find('meta',{'property':'og:title'}).get('content')

    user_score='-'
    try:
        user_score_1=results_page.find('div',{'class':'ratingValue'}).find('strong')
        user_score=user_score_1.find('span',{'itemprop':'ratingValue'}).get_text()
    except:
        pass
    
    ratingCount = '-'
    try:
        href_id='/title/tt'+imdbId+'/ratings?ref_=tt_ov_rt'
        ratingCount1=results_page.find('a',{'href':href_id})
        try:
            ratingCount=ratingCount1.find('span',{'itemprop':'ratingCount'}).get_text()
        except:
            pass
    except:
        pass
    
    budget = '-'
    for i in results_page.find_all('h4',{'class':'inline'}):
        if i.get_text()=='Budget:':
            b=i.parent.get_text()
            budget = b.split()[0].split(':')[1]
    
    revenue = "-"
    revenue_USA = '-'
    for i in results_page.find_all('h4',{'class':'inline'}):
        if i.get_text()=='Cumulative Worldwide Gross:':
            b=i.parent.get_text()
            revenue = b.split()[3]
            #revenue = b.split()[0].split(':')[1]
        
        if i.get_text()=='Gross USA:':
            p=i.parent.get_text()
            revenue_USA = p.split()[2]
    
    if revenue == "-":
        revenue = revenue_USA

    
    #<a href="/title/tt0008536/ratings?ref_=tt_ov_rt"><span class="small" itemprop="ratingCount">97</span></a>
    #soup.find('div',attrs={"class":"table-info"}).findAll('div')
    #type[2].find('strong').string 

    #featured_crew=[]
    #for i in results_page.find_all('li',{'class':'profile'}):
        #try:
            #featured_crew.append((i.find('a').get_text(),(i.find('a').get('href')).split('/')[2].split('-')[0],i.find('p',{'class':'character'}).get_text()))
        #except:
            #pass

    #top_billed_cast=[]
    #for i in results_page.find_all('li',{'class':'card'}):
        #try:
            #top_billed_cast.append((i.find('p').get_text(),i.find('a').get('href').split('/')[2].split('-')[0],i.find('p',{'class':'character'}).get_text()))
        #except:
            #pass

    #for i in results_page.find_all('strong'):
        #if i.get_text()=='Original Language':
            #original_language=i.parent.get_text().split()[2]

    #for i in results_page.find_all('strong'):
        #if i.get_text()=='Runtime':
            #if (i.parent.get_text().split()[1]) != "-":
                #runtime=(i.parent.get_text().split()[1])+(i.parent.get_text().split()[2])
            #else:
                #runtime="-"

    #for i in results_page.find_all('strong'):
        #if i.get_text()=='Budget':
            #budget=i.parent.get_text().split()[1]

    #for i in results_page.find_all('strong'):
        #if i.get_text()=='Revenue':
            #revenue=i.parent.get_text().split()[1]

    #genres=[]
    #for i in results_page.find('section',{'class':'genres right_column'}).find_all('a'):
        #genres.append(i.get_text())

    #all_features=[movieId,title,user_score,featured_crew,top_billed_cast,original_language,runtime,budget,revenue,genres]
    all_features=[movieId,title,user_score,ratingCount,budget,revenue]
    import pandas as pd
    pd.set_option('display.max_colwidth', -1)
    df=pd.DataFrame(all_features).transpose()
    #df.columns=['movieId','TITLE','USER SCORE','FEATURED CREW','TOP BILLED CAST','ORIGINAL LANGUAGE','RUN TIME','BUDGET','REVENUE','GENRE']
    df.columns=['movieId','TITLE','IMDB_SCORE','TOTAL_RATINGS','BUDGET','REVENUE']
    return df  
```
