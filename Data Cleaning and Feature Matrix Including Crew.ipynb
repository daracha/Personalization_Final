```python
def director_col(x):    #function creates a col with director_id
    a = x[1:-1]
    a1 = a.replace('(','|')
    a2 = a1.replace(')','|')
    a3 = a2.split('|')
    #print a3
    b = ''
    for l in a3:
        if 'Director' in l:            
            y=l.split(',')
            #print y
            b = y[1]
    return b    
        
    

#all_data['FEATURED CREW'].apply(director_col)
```

```python
all_data['Director'] = all_data['FEATURED CREW'].apply(director_col)
```

```python
def fcrew(x):    #function create a col with list of all crew_ids (except director)
    a = x[1:-1]
    a1 = a.replace('(','|')
    a2 = a1.replace(')','|')
    a3 = a2.split('|')
    b = []
    c=[]
    #fcrew1,fcrew2,fcrew3,fcrew4,fcrew5 = '-','-','-','-','-'
    p=1
    for l in a3:
        if 'Director' not in l:   
            #print l
            y=l.split(',')
            #print y
            try:
                b.append(y[1])
            except:
                pass
    j=0
    for i in range(len(b)):
        if b[i]!=' ':
            #print i,j
            c.append(b[i])
            j+=1
            
    for i in range(len(c),5):
        c.append('-')
    d=''
    for i in range(len(c)-1):
        d=d+str(c[i])+':'
    d=d+str(c[-1])
    return d

#all_data['FEATURED CREW'].apply(fcrew)
```

```python
all_data['fcrew_id']= all_data['FEATURED CREW'].apply(fcrew)
```

```python
def top_cast(x):    #function create a col with list of all top_cast_ids (in decreasing order of remuneration)
    a = x[1:-1]
    a1 = a.replace('(','|')
    a2 = a1.replace(')','|')
    a3 = a2.split('|')
    b = []
    c=[]
    p=1
    for l in a3: 
        #print l
        y=l.split(',')
        #print y
        try:
            b.append(y[1])
        except:
            pass
    j=0
    for i in range(len(b)):
        if b[i]!=' ':
            #print i,j
            c.append(b[i])
            
    for i in range(len(c),5):
        c.append('-')
    d=''
    for i in range(len(c)-1):
        d=d+str(c[i])+':'
    d=d+str(c[-1])
    return d
```

```python
all_data['top_cast_ids']= all_data['TOP BILLED CAST'].apply(top_cast)
```

```python
def change_time(x):  #convert h,mins into only mins for run-time
    h=x[0]
    m=x[2:-1]
    time = 0
    #print("hour is",h,"min is",m)
    try:
        time = 60*int(h) + int(m)
    #except ValueError:
        #print("hour is",h,"min is",m)
    except:
        pass
    return time
```

```python
def genre_split(x):
    a = x.split("[")[1].split(']')[0].split(',')
    if len(a)<=10:
        for i in range(len(a),10):
            a.append('-')
    else:
        a=a[:10]
    return a
```

```python
pop_class=[]
for i in (range(1,11)):
    a='pop_class'+str(i)
    pop_class.append(a)
print pop_class
```

```python
#convert string formatted TOTAL_RATINGS to int format
def str2int(x):
    #print type(x)
        a =x
        if a!='-':    
            b=a.split(',')
            p=''
            for i in b:
                p=p+i
            #print p,type(p)
            #p=float(p)
            return int(p)
```

```python
all_data['ratings_count']=all_data['TOTAL_RATINGS'].apply(str2int) #ratings_count = no. of user ratings in IMDb
```

```python
all_data['popular_class'] = pd.qcut(all_data['ratings_count'], 10, labels=['pop_class1', 'pop_class2', 'pop_class3', 'pop_class4', 'pop_class5', 'pop_class6', 'pop_class7', 'pop_class8', 'pop_class9', 'pop_class10'])
```
