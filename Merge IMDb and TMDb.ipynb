```python
import pandas as pd
import numpy as np
```

```python
#reading files
tmdb = pd.read_csv('TMDb_scraping_edited.csv')
imdb = pd.read_csv('IMDb_scraped_100k_data_edited.csv')
```

```python
merged_data = pd.merge(tmdb, imdb, on='movieId', how='inner',suffixes=('_tmdb', '_imdb'))
```

```python
merged_data
```

```python
merged_data.shape[0]
```

```python
merged_data['BUDGET']='-'
```

```python
for i in (range(merged_data.shape[0])): #append tmdb budget col with imdb budget col #structure needs improvement
    if merged_data['BUDGET_tmdb'][i] != '-':
        merged_data['BUDGET'][i] = merged_data['BUDGET_tmdb'][i]
    elif merged_data['BUDGET_imdb'][i] != '-':
        merged_data['BUDGET'][i] = merged_data['BUDGET_imdb'][i]
    #else:
        #merged_data['BUDGET'][i] = '-'
        
```

```python
merged_data['REVENUE']='-'
```

```python
for i in (range(merged_data.shape[0])): #append tmdb rev col with imdb rev col 
    if merged_data['REVENUE_tmdb'][i] != '-':
        merged_data['REVENUE'][i] = merged_data['REVENUE_tmdb'][i]
    elif merged_data['REVENUE_imdb'][i] != '-':
        merged_data['REVENUE'][i] = merged_data['REVENUE_imdb'][i]
```

```python
del merged_data['BUDGET_tmdb']
del merged_data['REVENUE_tmdb']
del merged_data['BUDGET_imdb']
del merged_data['REVENUE_imdb']
del merged_data['TITLE_imdb']
```

```python
merged_data = merged_data.rename(index=str, columns={"TITLE_tmdb": "TITLE"})
```

```python
merged_data = merged_data.rename(index=str, columns={"USER SCORE": "TMDB_SCORE"})
```
